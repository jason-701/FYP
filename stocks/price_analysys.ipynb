```python
# start to nov 2001 (dot com crash)
# not bad
# df = df.iloc[:2228]

# nov 01 to june 2009
# not bad
# df = df.iloc[2228:4135]

# june 2009 to april 2020
# ass
# df = df.iloc[4135:6862]

df_section_1 = df.iloc[:2228]
df_section_2 = df.iloc[2228:4135]
df_section_3 = df.iloc[4135:6862]
```

```python
# Adding SMA as part of the feature test
short_term_period = 20
medium_term_period = 50
long_term_period = 200

# Compute the SMAs
df_section_1['SMA_20'] = df_section_1['Close'].rolling(window=short_term_period).mean()
df_section_1['SMA_50'] = df_section_1['Close'].rolling(window=medium_term_period).mean()
df_section_1['SMA_200'] = df_section_1['Close'].rolling(window=long_term_period).mean()

df_section_1['EMA_10'] = df_section_1['Close'].ewm(span=10, adjust=False).mean()  # Short-term EMA
df_section_1['EMA_50'] = df_section_1['Close'].ewm(span=50, adjust=False).mean()  # Medium-term EMA
df_section_1['EMA_200'] = df_section_1['Close'].ewm(span=200, adjust=False).mean()  # Long-term EMA    

df_section_1 = df_section_1.dropna()
```

```python
# Adding SMA as part of the feature test
short_term_period = 20
medium_term_period = 50
long_term_period = 200

# Compute the SMAs
df_section_2['SMA_20'] = df_section_2['Close'].rolling(window=short_term_period).mean()
df_section_2['SMA_50'] = df_section_2['Close'].rolling(window=medium_term_period).mean()
df_section_2['SMA_200'] = df_section_2['Close'].rolling(window=long_term_period).mean()

df_section_2['EMA_10'] = df_section_2['Close'].ewm(span=10, adjust=False).mean()  # Short-term EMA
df_section_2['EMA_50'] = df_section_2['Close'].ewm(span=50, adjust=False).mean()  # Medium-term EMA
df_section_2['EMA_200'] = df_section_2['Close'].ewm(span=200, adjust=False).mean()  # Long-term EMA    

df_section_2 = df_section_2.dropna()
```

```python
# Adding SMA as part of the feature test
short_term_period = 20
medium_term_period = 50
long_term_period = 200

# Compute the SMAs
df_section_3['SMA_20'] = df_section_3['Close'].rolling(window=short_term_period).mean()
df_section_3['SMA_50'] = df_section_3['Close'].rolling(window=medium_term_period).mean()
df_section_3['SMA_200'] = df_section_3['Close'].rolling(window=long_term_period).mean()

df_section_3['EMA_10'] = df_section_3['Close'].ewm(span=10, adjust=False).mean()  # Short-term EMA
df_section_3['EMA_50'] = df_section_3['Close'].ewm(span=50, adjust=False).mean()  # Medium-term EMA
df_section_3['EMA_200'] = df_section_3['Close'].ewm(span=200, adjust=False).mean()  # Long-term EMA    

df_section_3 = df_section_3.dropna()
```

```python
feature_columns = ["Open", "High", "Low", "Close", "Volume", "SMA_20", "SMA_50", "SMA_200", "EMA_10", "EMA_50", "EMA_200"]
```

```python

X_1 = df_section_1[feature_columns]

# Split data
train_size = int(X_1.shape[0] * 0.7)
val_size = int(X_1.shape[0] * 0.15)

# Initialize scaler
# scaler = RobustScaler()
scaler = MinMaxScaler()
# scaler = RobustScaler()

# Split data into train, validation, and test
temp_train = X_1[:train_size]
temp_val = X_1[train_size:train_size + val_size]
temp_test = X_1[train_size + val_size:]

temp_train = temp_train[feature_columns]
temp_val = temp_val[feature_columns]
temp_test = temp_test[feature_columns]

# Fit and transform for train, transform for validation and test
temp_train = scaler.fit_transform(temp_train)
temp_val = scaler.transform(temp_val)
temp_test = scaler.transform(temp_test)

# Convert NumPy arrays back to DataFrames
temp_train = pd.DataFrame(temp_train, columns=feature_columns)
temp_val = pd.DataFrame(temp_val, columns=feature_columns)
temp_test = pd.DataFrame(temp_test, columns=feature_columns)

# Concatenate the data
X_scaled1 = pd.concat([temp_train, temp_val, temp_test], axis=0)

# Split sequentially
X_train1 = X_scaled1[:train_size]
X_val1 = X_scaled1[train_size:train_size + val_size]
X_test1 = X_scaled1[train_size + val_size:]

# Print sizes to confirm
print(f"Train: {len(X_train1)}, Val: {len(X_val1)}, Test: {len(X_test1)}")
```

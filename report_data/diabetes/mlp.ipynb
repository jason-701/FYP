```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('mlp.csv')
```

```python
df.describe()
```

```python
metrics_high = ["accuracy", "precision"]
metrics_f1 = ["f1_score"]
metrics_low = ["recall"]

# Define color palettes
colors_high = ["#1f77b4", "#ff7f0e"]  # Blue, Orange
colors_f1 = ["#2ca02c"]  # Green
colors_low = ["#d62728"]  # Red

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), gridspec_kw={'width_ratios': [2, 1, 1]})

# Boxplot for Accuracy and Precision
sns.boxplot(data=df[metrics_high], palette=colors_high, ax=axes[0])
axes[0].set_title("Accuracy and Precision", fontsize=12, fontweight='bold')
axes[0].set_ylabel("Score", fontsize=10)
axes[0].set_xticklabels(metrics_high, fontsize=10)
axes[0].grid(True, linestyle="--", alpha=0.6)

# Boxplot for F1 Score
sns.boxplot(data=df[metrics_f1], palette=colors_f1, ax=axes[1])
axes[1].set_title("F1 Score", fontsize=12, fontweight='bold')
axes[1].set_ylabel("Score", fontsize=10)
axes[1].set_xticklabels(metrics_f1, fontsize=10)
axes[1].grid(True, linestyle="--", alpha=0.6)

# Boxplot for Recall
sns.boxplot(data=df[metrics_low], palette=colors_low, ax=axes[2])
axes[2].set_title("Recall", fontsize=12, fontweight='bold')
axes[2].set_ylabel("Score", fontsize=10)
axes[2].set_xticklabels(metrics_low, fontsize=10)
axes[2].grid(True, linestyle="--", alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Separate metrics
metrics = ["accuracy", "precision", "recall", "f1_score"]
colors = ["#1f77b4", "#ff7f0e", "#d62728", "#2ca02c"]  # Blue, Orange, Red, Green

# Create subplots
fig, axes = plt.subplots(1, 4, figsize=(14, 10))
axes = axes.flatten()

for i, metric in enumerate(metrics):
    sns.violinplot(data=df[metric], color=colors[i], ax=axes[i], inner=None, alpha=0.5)
    sns.boxplot(data=df[metric], color=colors[i], ax=axes[i], width=0.2, boxprops={'zorder': 2})
    axes[i].set_title(metric, fontsize=12, fontweight='bold')
    axes[i].set_ylabel("Score", fontsize=10)
    axes[i].grid(True, linestyle="--", alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Melt the dataframe for better visualization
df_melted = df.melt(id_vars=["trial"], var_name="Metric", value_name="Score")

# Define color palette
palette = {"accuracy": "#1f77b4", "precision": "#ff7f0e", "recall": "#d62728", "f1_score": "#2ca02c"}

# Create a combined violin and box plot
plt.figure(figsize=(12, 6))
sns.violinplot(x="Metric", y="Score", data=df_melted, palette=palette, inner=None, alpha=0.5)
sns.boxplot(x="Metric", y="Score", data=df_melted, width=0.2, palette=palette, boxprops={'zorder': 2})

# Title and labels
plt.title("Combined Violin and Box Plot of Model Metrics", fontsize=14, fontweight='bold')
plt.ylabel("Score", fontsize=12)
plt.xlabel("Metric", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

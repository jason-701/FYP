```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv("mlp_improved_multi_trials.csv")  # Replace with actual file path

# Extract confusion matrix values from the first row and convert to integers
TN, FP, FN, TP = map(int, df.iloc[0][["Original TN", "Original FP", "Original FN", "Original TP"]])

# Create confusion matrix
conf_matrix = np.array([[TN, FP], [FN, TP]])

# Define class labels
labels = ["No Diabetes", "Diabetes"]

# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix (Original Model)")

plt.show()
```

```python
# Normalize confusion matrix by row (i.e., percentage of actual class)
conf_matrix_norm = conf_matrix.astype('float') / conf_matrix.sum(axis=1, keepdims=True)

# Plot normalized heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_norm, annot=True, fmt=".2%", cmap="Blues", xticklabels=labels, yticklabels=labels)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Normalized Confusion Matrix (Original Model)")

plt.show()
```

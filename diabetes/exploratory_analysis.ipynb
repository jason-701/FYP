```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 16))

# Distribution of Target Variable (Diabetes)
sns.countplot(x='diabetes', data=data, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Diabetes (target variable)')
axes[0, 0].set_xlabel('Diabetes')
axes[0, 0].set_ylabel('Count')

# Distribution of Age
sns.histplot(data['age'], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Age Distribution')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Frequency')

# Gender Distribution
sns.countplot(x='gender', data=data, ax=axes[0, 2])
axes[0, 2].set_title('Gender Distribution')
axes[0, 2].set_xlabel('Gender')
axes[0, 2].set_ylabel('Count')

# BMI Distribution
sns.histplot(data['bmi'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('BMI Distribution')
axes[1, 0].set_xlabel('BMI')
axes[1, 0].set_ylabel('Frequency')

# Distribution of HbA1c Level
sns.histplot(data['HbA1c_level'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('HbA1c Level Distribution')
axes[1, 1].set_xlabel('HbA1c Level')
axes[1, 1].set_ylabel('Frequency')

# Distribution of Blood Glucose Level
sns.histplot(data['blood_glucose_level'], bins=30, kde=True, ax=axes[1, 2])
axes[1, 2].set_title('Blood Glucose Level Distribution')
axes[1, 2].set_xlabel('Blood Glucose Level')
axes[1, 2].set_ylabel('Frequency')

# Smoking History Distribution
sns.countplot(x='smoking_history', data=data, ax=axes[2, 0])
axes[2, 0].set_title('Smoking History Distribution')
axes[2, 0].set_xlabel('Smoking History')
axes[2, 0].set_ylabel('Count')
axes[2, 0].set_xticklabels(axes[2, 0].get_xticklabels(), rotation=45)

# Hypertension and Heart Disease Distribution
sns.countplot(x='hypertension', data=data, ax=axes[2, 1])
axes[2, 1].set_title('Hypertension Distribution')
axes[2, 1].set_xlabel('Hypertension')
axes[2, 1].set_ylabel('Count')

sns.countplot(x='heart_disease', data=data, ax=axes[2, 2])
axes[2, 2].set_title('Heart Disease Distribution')
axes[2, 2].set_xlabel('Heart Disease')
axes[2, 2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
